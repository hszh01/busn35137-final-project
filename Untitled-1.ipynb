```python
# ============================================================
# Asset Pricing: XGBoost + MLP Models
# Cross-Sectional Stock Return Prediction
# ============================================================

# ============================================================
# BLOCK 1: Import Libraries
# ============================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

# XGBoost
import xgboost as xgb
```

```python
# PyTorch (MLP)
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# Evaluation & Tuning
from sklearn.metrics import mean_squared_error
import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)

# SHAP (MLP interpretability)
import shap

print("All libraries loaded successfully.")
print(f"PyTorch version: {torch.__version__}")
print(f"XGBoost version: {xgb.__version__}")
```

```python
# ============================================================
# BLOCK 2: Load and Split Data
# ============================================================

# --- Load your dataset ---
# Replace the path below with your actual file path
df = pd.read_parquet("/Users/duanfuzhen/Desktop/model_input_2001_2024.parquet")

# --- Define feature columns and target ---
FEATURES = [
    'mktcap_z', 'log_ret_z', 'momentum_z',
    'volatility_z', 'beta_z', 'book_to_market_z',
    'roa_z', 'investment_z', 'ceq_z', 'ni_z', 'at_z'
]
TARGET = 'ret_fwd'
```

```python
print(df['split'].value_counts())
```

```python
# Winsorize ret_fwd to remove extreme outliers
# Cap returns at 1st and 99th percentile
low  = df['ret_fwd'].quantile(0.01)
high = df['ret_fwd'].quantile(0.99)
df['ret_fwd'] = df['ret_fwd'].clip(lower=low, upper=high)

print(f"Return range after winsorize: [{low:.4f}, {high:.4f}]")
print(f"New skewness: {df['ret_fwd'].skew():.4f}")



# --- Split using the pre-defined 'split' column ---
train_df = df[df['split'] == 'train'].copy()
val_df   = df[df['split'] == 'val'].copy()
test_df  = df[df['split'] == 'test'].copy()

X_train, y_train = train_df[FEATURES].values, train_df[TARGET].values
X_val,   y_val   = val_df[FEATURES].values,   val_df[TARGET].values
X_test,  y_test  = test_df[FEATURES].values,  test_df[TARGET].values

print(f"Train size : {X_train.shape[0]:,}")
print(f"Val size   : {X_val.shape[0]:,}")
print(f"Test size  : {X_test.shape[0]:,}")

# --- Check time range of each split ---
for split in ['train', 'val', 'test']:
    sub = df[df['split'] == split]
    print(f"{split}: {sub['month'].min()} → {sub['month'].max()}")
```

```python
# ============================================================
# BLOCK 3: Evaluation Helper
# ============================================================

def oos_r2(y_true, y_pred):
    """
    Out-of-sample R² following Gu, Kelly, Xiu (2020).
    Benchmark is the historical mean return (not zero).
    A positive value means the model beats the mean benchmark.
    """
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - ss_res / ss_tot

def evaluate(y_true, y_pred, model_name="Model"):
    r2   = oos_r2(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"[{model_name}]  OOS R² = {r2:.4f}   RMSE = {rmse:.4f}")
    return {"model": model_name, "oos_r2": r2, "rmse": rmse}
```

```python
# ============================================================
# BLOCK 4: Linear Regression Baseline
# ============================================================
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

results = []
results.append(evaluate(y_test, y_pred_ridge, "Ridge (Baseline)"))
```

```python
# ============================================================
# BLOCK 5: XGBoost — Hyperparameter Tuning with Optuna
# ============================================================

def xgb_objective(trial):
    """
    Optuna objective: train XGBoost with suggested params,
    return validation RMSE (lower is better).
    """
    params = {
        "n_estimators"     : trial.suggest_int("n_estimators", 100, 600),
        "max_depth"        : trial.suggest_int("max_depth", 3, 7),
        "learning_rate"    : trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
        "subsample"        : trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree" : trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "reg_lambda"       : trial.suggest_float("reg_lambda", 1e-3, 10.0, log=True),
        "objective"        : "reg:squarederror",
        "eval_metric"      : "rmse",
        "random_state"     : 42,
        "n_jobs"           : -1,
    }
    model = xgb.XGBRegressor(**params, early_stopping_rounds=20, verbosity=0)
    model.fit(X_train, y_train,
              eval_set=[(X_val, y_val)],
              verbose=False)
    y_pred = model.predict(X_val)
    return np.sqrt(mean_squared_error(y_val, y_pred))

print("Tuning XGBoost... (this may take a few minutes)")
study_xgb = optuna.create_study(direction="minimize")
study_xgb.optimize(xgb_objective, n_trials=30)

best_xgb_params = study_xgb.best_params
best_xgb_params.update({
    "objective": "reg:squarederror",
    "random_state": 42,
    "n_jobs": -1,
})
print("Best XGBoost params:", best_xgb_params)
```

```python
# ============================================================
# BLOCK 6: XGBoost Final Model (Manual optimized params)
xgb_model = xgb.XGBRegressor(
    n_estimators          = 500,
    max_depth             = 4,
    learning_rate         = 0.05,
    subsample             = 0.7,
    colsample_bytree      = 0.7,
    reg_lambda            = 5.0,
    min_child_weight      = 50,
    objective             = "reg:squarederror",
    early_stopping_rounds = 30,
    random_state          = 42,
    n_jobs                = -1,
    verbosity             = 0
)

xgb_model.fit(X_train, y_train,
              eval_set=[(X_val, y_val)],
              verbose=False)

y_pred_xgb = xgb_model.predict(X_test)
results = []
results.append(evaluate(y_test, y_pred_ridge, "Ridge (Baseline)"))
results.append(evaluate(y_test, y_pred_xgb, "XGBoost"))
```

```python
# ============================================================
# BLOCK 7: XGBoost — Feature Importance Plot
# ============================================================

importance_df = pd.DataFrame({
    "feature"   : FEATURES,
    "importance": xgb_model.feature_importances_
}).sort_values("importance", ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))
ax.barh(importance_df["feature"], importance_df["importance"], color="steelblue")
ax.set_xlabel("Feature Importance (Gain)", fontsize=12)
ax.set_title("XGBoost: Feature Importance", fontsize=13, fontweight="bold")
ax.axvline(importance_df["importance"].mean(), color="red",
           linestyle="--", label="Mean importance")
ax.legend()
plt.tight_layout()
plt.savefig("xgb_feature_importance.png", dpi=150)
plt.show()
print("Saved: xgb_feature_importance.png")
```

```python
# ============================================================
# BLOCK 8: XGBoost — Rolling Feature Importance (Over Time)
# This shows which features matter in different market regimes.
# ============================================================

# Divide test period into 4 sub-periods
test_df = test_df.copy()
test_df["quarter"] = pd.qcut(pd.to_datetime(test_df["month"]), q=4, labels=False)

fig, axes = plt.subplots(1, 4, figsize=(18, 5), sharey=True)
colors = ["steelblue", "darkorange", "seagreen", "firebrick"]

for q in range(4):
    sub = test_df[test_df["quarter"] == q]
    X_sub = sub[FEATURES].values
    imp = xgb_model.get_booster().get_score(importance_type='gain')
    # Use SHAP values for sub-period importance
    explainer_sub = shap.TreeExplainer(xgb_model)
    shap_vals_sub = explainer_sub.shap_values(X_sub[:2000])  # sample for speed
    mean_abs_shap = np.abs(shap_vals_sub).mean(axis=0)

    imp_df = pd.DataFrame({"feature": FEATURES, "importance": mean_abs_shap})
    imp_df = imp_df.sort_values("importance", ascending=True)

    axes[q].barh(imp_df["feature"], imp_df["importance"], color=colors[q])
    axes[q].set_title(f"Sub-period {q+1}", fontweight="bold")
    axes[q].set_xlabel("Mean |SHAP|")

fig.suptitle("XGBoost: Feature Importance Across Time Periods", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("xgb_rolling_importance.png", dpi=150)
plt.show()
print("Saved: xgb_rolling_importance.png")
```

```python
# ============================================================
# BLOCK 9: MLP — Model Architecture
# ============================================================

class MLP(nn.Module):
    """
    Feedforward Neural Network for return prediction.
    Architecture: Input → [256 → 128 → 64] → Output
    Each hidden layer: Linear + BatchNorm + ReLU + Dropout
    """
    def __init__(self, input_dim, hidden_dims, dropout_rate):
        super(MLP, self).__init__()
        layers = []
        prev_dim = input_dim
        for h_dim in hidden_dims:
            layers += [
                nn.Linear(prev_dim, h_dim),
                nn.BatchNorm1d(h_dim),
                nn.ReLU(),
                nn.Dropout(dropout_rate)
            ]
            prev_dim = h_dim
        layers.append(nn.Linear(prev_dim, 1))   # output: predicted return
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x).squeeze(-1)
```

```python
# ============================================================
# BLOCK 10: MLP — Hyperparameter Tuning with Optuna
# ============================================================

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

def mlp_objective(trial):
    """
    Optuna objective for MLP: returns validation RMSE.
    Searches over architecture and training hyperparams.
    """
    # Architecture
    n_layers     = trial.suggest_int("n_layers", 2, 4)
    hidden_dims  = [trial.suggest_categorical(
                        f"h_{i}", [64, 128, 256, 512]) for i in range(n_layers)]
    dropout_rate = trial.suggest_float("dropout", 0.1, 0.5)

    # Training
    lr         = trial.suggest_float("lr", 1e-4, 1e-2, log=True)
    batch_size = trial.suggest_categorical("batch_size", [512, 1024, 2048])
    weight_decay = trial.suggest_float("weight_decay", 1e-5, 1e-2, log=True)

    # Build data loaders
    X_tr = torch.FloatTensor(X_train)
    y_tr = torch.FloatTensor(y_train)
    X_vl = torch.FloatTensor(X_val).to(DEVICE)
    y_vl = torch.FloatTensor(y_val).to(DEVICE)

    loader = DataLoader(TensorDataset(X_tr, y_tr),
                        batch_size=batch_size, shuffle=True)

    # Build model
    model = MLP(len(FEATURES), hidden_dims, dropout_rate).to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    criterion = nn.MSELoss()

    # Train for up to 30 epochs with early stopping
    best_val_loss = float('inf')
    patience, patience_counter = 5, 0

    for epoch in range(30):
        model.train()
        for X_batch, y_batch in loader:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            loss = criterion(model(X_batch), y_batch)
            loss.backward()
            optimizer.step()

        model.eval()
        with torch.no_grad():
            val_loss = criterion(model(X_vl), y_vl).item()

        if val_loss < best_val_loss - 1e-6:
            best_val_loss = val_loss
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                break

    return np.sqrt(best_val_loss)

print("Tuning MLP... (this may take several minutes)")
study_mlp = optuna.create_study(direction="minimize")
study_mlp.optimize(mlp_objective, n_trials=20)

best_mlp_params = study_mlp.best_params
print("Best MLP params:", best_mlp_params)
```

```python
# ============================================================
# BLOCK 11: MLP — Final Training with Best Params
# ============================================================

# Reconstruct best architecture
n_layers = best_mlp_params["n_layers"]
hidden_dims = [best_mlp_params[f"h_{i}"] for i in range(n_layers)]
dropout_rate = best_mlp_params["dropout"]
lr = best_mlp_params["lr"]
batch_size = best_mlp_params["batch_size"]
weight_decay = best_mlp_params["weight_decay"]

# Build final model
mlp_model = MLP(len(FEATURES), hidden_dims, dropout_rate).to(DEVICE)
optimizer = torch.optim.Adam(mlp_model.parameters(), lr=lr, weight_decay=weight_decay)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
criterion = nn.MSELoss()

X_tr_t = torch.FloatTensor(X_train)
y_tr_t = torch.FloatTensor(y_train)
X_vl_t = torch.FloatTensor(X_val).to(DEVICE)
y_vl_t = torch.FloatTensor(y_val).to(DEVICE)
X_te_t = torch.FloatTensor(X_test).to(DEVICE)

loader = DataLoader(TensorDataset(X_tr_t, y_tr_t),
                    batch_size=batch_size, shuffle=True)

# Training loop with early stopping
best_val_loss = float('inf')
patience_counter = 0
PATIENCE = 10
train_losses, val_losses = [], []
best_state = None

print("Training MLP...")
for epoch in range(100):
    mlp_model.train()
    epoch_loss = 0
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
        optimizer.zero_grad()
        loss = criterion(mlp_model(X_batch), y_batch)
        loss.backward()
        nn.utils.clip_grad_norm_(mlp_model.parameters(), max_norm=1.0)
        optimizer.step()
        epoch_loss += loss.item()

    mlp_model.eval()
    with torch.no_grad():
        val_loss = criterion(mlp_model(X_vl_t), y_vl_t).item()

    train_losses.append(epoch_loss / len(loader))
    val_losses.append(val_loss)
    scheduler.step(val_loss)

    if val_loss < best_val_loss - 1e-6:
        best_val_loss = val_loss
        best_state = {k: v.cpu().clone() for k, v in mlp_model.state_dict().items()}
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= PATIENCE:
            print(f"Early stopping at epoch {epoch+1}")
            break

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d} | Train Loss: {train_losses[-1]:.5f} | Val Loss: {val_loss:.5f}")

# Load best weights
mlp_model.load_state_dict(best_state)

# Final test evaluation
mlp_model.eval()
with torch.no_grad():
    y_pred_mlp = mlp_model(X_te_t).cpu().numpy()

results.append(evaluate(y_test, y_pred_mlp, "MLP"))
```

```python
# ============================================================
# BLOCK 13: MLP — SHAP Analysis (Fixed)
# ============================================================

# Wrapper to make MLP output 2D (required by SHAP)
class MLPWrapper(nn.Module):
    def __init__(self, model):
        super().__init__()
        self.model = model
    def forward(self, x):
        return self.model(x).unsqueeze(1)  # (batch,) → (batch, 1)

wrapped_model = MLPWrapper(mlp_model).to(DEVICE)
wrapped_model.eval()

background  = torch.FloatTensor(X_train[:200]).to(DEVICE)
test_sample = torch.FloatTensor(X_test[:500]).to(DEVICE)

explainer_mlp = shap.GradientExplainer(wrapped_model, background)
shap_values   = explainer_mlp.shap_values(test_sample)

# Convert to 2D numpy array
if isinstance(shap_values, list):
    shap_values = shap_values[0]
shap_values = np.array(shap_values)
if shap_values.ndim == 3:
    shap_values = shap_values[:, :, 0]

print(f"SHAP values shape: {shap_values.shape}")  # Should be (500, 11)

# SHAP Summary Plot
shap.summary_plot(shap_values,
                  X_test[:500],
                  feature_names=FEATURES,
                  show=False)
plt.title("MLP SHAP Summary Plot", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("mlp_shap_summary.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved: mlp_shap_summary.png")

# SHAP Dependence Plot for most important feature
mean_abs        = np.abs(shap_values).mean(axis=0)
top_feature_idx = np.argmax(mean_abs)
top_feature     = FEATURES[top_feature_idx]

fig, ax = plt.subplots(figsize=(7, 5))
shap.dependence_plot(top_feature_idx,
                     shap_values,
                     X_test[:500],
                     feature_names=FEATURES,
                     ax=ax,
                     show=False)
ax.set_title(f"MLP SHAP Dependence: {top_feature}", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("mlp_shap_dependence.png", dpi=150, bbox_inches='tight')
plt.show()
print(f"Saved: mlp_shap_dependence.png  (top feature: {top_feature})")
```

```python
# ============================================================
# BLOCK 14: Long-Short Portfolio — Economic Evaluation
# ============================================================

def long_short_portfolio(df_split, y_pred, model_name, top_pct=0.1):
    """
    Each month: buy top 10% predicted return stocks (long),
    sell bottom 10% (short). Compute equal-weighted monthly return.
    """
    df_split = df_split.copy()
    df_split['pred'] = y_pred
    df_split['month'] = pd.to_datetime(df_split['month'])

    monthly_returns = []
    for month, group in df_split.groupby('month'):
        n = len(group)
        if n < 20:
            continue
        k = max(1, int(n * top_pct))
        long_ret  = group.nlargest(k, 'pred')['ret_fwd'].mean()
        short_ret = group.nsmallest(k, 'pred')['ret_fwd'].mean()
        monthly_returns.append({
            'month': month,
            'ls_return': long_ret - short_ret
        })

    port_df = pd.DataFrame(monthly_returns).set_index('month')
    cumret = (1 + port_df['ls_return']).cumprod()
    annual_ret = port_df['ls_return'].mean() * 12
    sharpe = (port_df['ls_return'].mean() / port_df['ls_return'].std()) * np.sqrt(12)

    print(f"[{model_name}] Annualized Return: {annual_ret:.2%}  |  Sharpe Ratio: {sharpe:.2f}")
    return port_df, cumret, {"model": model_name, "annual_ret": annual_ret, "sharpe": sharpe}

# Build portfolios for each model
port_ridge, cum_ridge, eco_ridge = long_short_portfolio(test_df, y_pred_ridge, "Ridge")
port_xgb,   cum_xgb,   eco_xgb   = long_short_portfolio(test_df, y_pred_xgb,   "XGBoost")
port_mlp,   cum_mlp,   eco_mlp   = long_short_portfolio(test_df, y_pred_mlp,   "MLP")

# Plot cumulative returns
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(cum_ridge, label="Ridge (Baseline)", color="gray", linestyle="--")
ax.plot(cum_xgb,   label="XGBoost",          color="steelblue")
ax.plot(cum_mlp,   label="MLP",              color="darkorange")
ax.axhline(1.0, color="black", linewidth=0.8, linestyle=":")
ax.set_xlabel("Date")
ax.set_ylabel("Cumulative Return (Long-Short)")
ax.set_title("Long-Short Portfolio: Cumulative Returns", fontsize=13, fontweight="bold")
ax.legend()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.tight_layout()
plt.savefig("portfolio_cumulative_returns.png", dpi=150)
plt.show()
print("Saved: portfolio_cumulative_returns.png")
```
